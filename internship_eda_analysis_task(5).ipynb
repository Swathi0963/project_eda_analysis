```python
# List of specializations to check
specializations_of_interest = [
    'Programming Analyst', 
    'Software Engineer', 
    'Hardware Engineer', 
    'Associate Engineer'
]

# Count the number of Computer Science graduates earning within the specified salary range
salary_range_cs = df[
    (df['salary'] >= 250000) & (df['salary'] <= 300000) &
    (df['specialization'].isin(specializations_of_interest)) &
    (df['specialization'] == 'Computer Science')  # Assuming there's a column named 'branch' indicating the student's field
]

count_within_range_cs = salary_range_cs.shape[0]

print(f'Number of Computer Science graduates earning between 2.5 to 3 lakhs in specified specializations: {count_within_range_cs}')
```

```python
from scipy.stats import chi2_contingency

# Cross-tabulation of Gender and Specialization
gender_specialization_crosstab = pd.crosstab(df['gender'], df['specialization'])
print(gender_specialization_crosstab)


# Performing Chi-square test
chi2, p_value, dof, expected = chi2_contingency(gender_specialization_crosstab)

# Displaying the results
print(f"Chi-square statistic: {chi2}")
print(f"p-value: {p_value}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:")
print(expected)

# Checking significance
alpha = 0.05  # 95% confidence level
if p_value < alpha:
    print("The p-value is less than 0.05, so we reject the null hypothesis. There is a significant relationship between gender and specialization.")
else:
    print("The p-value is greater than 0.05, so we fail to reject the null hypothesis. There is no significant relationship between gender and specialization.")
```
